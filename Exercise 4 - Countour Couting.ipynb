```python
# Exercise 4, count the baloon in the picture
import cv2
import imutils
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('sample.png')
plt.imshow(img, ); plt.axis("off");
```

```python
gray  = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, ); plt.axis("off");
```

```python
thres = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV,21,10)
plt.imshow(thres, ); plt.axis("off");
```

```python
contours = cv2.findContours(thres, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# import reprlib
# print(reprlib.repr(contours))
len(contours)
```

```python
contours = imutils.grab_contours(contours)
```

```python
len(contours)
```

```python
contours = sorted(contours, key=cv2.contourArea, reverse=False)
```

```python
len(contours)
```

```python

number = 0
# loop over our contours
for c in contours:
    (x, y, w, h) = cv2.boundingRect(c)
    #print(x, y, w, h)
    #cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    # approximate the contour
    if (85<w<150) and(100<h<150):
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
        number +=1

print("Number of Contours found = " + str(number))
#cv2.drawContours(img, contours, -1, (0, 255, 0), 3)
```
